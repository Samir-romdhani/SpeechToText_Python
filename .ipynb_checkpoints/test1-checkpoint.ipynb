```python
import os
import speech_recognition as sr
from tqdm import tqdm
import IPython.display as ipd
import os
import pandas as pd
import librosa as lb
import glob 

import urllib3.contrib.pyopenssl

with open("api-key.json") as f:
    GOOGLE_CLOUD_SPEECH_CREDENTIALS = f.read()

#print(GOOGLE_CLOUD_SPEECH_CREDENTIALS)

r = sr.Recognizer()
files = os.listdir('source1/')

all_text = []

for f in tqdm(files):
    name = "source1/" + f
    # Load audio file
    #print(name)
    ipd.Audio(name)
    data, sampling_rate = lb.load(name)
    plt.figure(figsize=(8, 2))
    lb.display.waveplot(data, sr=sampling_rate)
    
    with sr.AudioFile(name) as source:
        audio = r.record(source)
        text = r.recognize_google(audio, language="us")
        #text = r.recognize_google_cloud(audio, credentials_json=GOOGLE_CLOUD_SPEECH_CREDENTIALS)
        all_text.append(text)

transcript = ""
for i, t in enumerate(all_text):
    total_seconds = i * 30
    # Cool shortcut from:text = r.recognize_google(audio, language="tn")
    # https://stackoverflow.com/questions/775049/python-time-seconds-to-hms
    # to get hours, minutes and seconds
    m, s = divmod(total_seconds, 60)
    h, m = divmod(m, 60)

    # Format time as h:m:s - 30 seconds of text
    transcript = transcript + "{:0>2d}:{:0>2d}:{:0>2d} {}\n".format(h, m, s, t)

print(transcript)

with open("transcripts/transcript.txt", "w") as f:
    f.write(transcript)
```

```python
import os
import speech_recognition as sr
from tqdm import tqdm
from multiprocessing.dummy import Pool
import IPython.display as ipd
pool = Pool(8) # Number of concurrent threads

with open("api-key.json") as f:
    GOOGLE_CLOUD_SPEECH_CREDENTIALS = f.read()

r = sr.Recognizer()
files = os.listdir('source1/')

def transcribe(data):
    idx, file = data
    name = "source1/" + file
    print(name + " started")
    # Load audio file
    with sr.AudioFile(name) as source:
        audio = r.record(source)
    # Transcribe audio file
    #text = r.recognize_google_cloud(audio, credentials_json=GOOGLE_CLOUD_SPEECH_CREDENTIALS)
    text = r.recognize_google(audio, language="us")
    print(name + " done")
    #ipd.Audio(name)
    return {
        "idx": idx,
        "text": text
    }

all_text = pool.map(transcribe, enumerate(files))
pool.close()
pool.join()

transcript = ""
for t in sorted(all_text, key=lambda x: x['idx']):
    total_seconds = t['idx'] * 30
    # Cool shortcut from:
    # https://stackoverflow.com/questions/775049/python-time-seconds-to-hms
    # to get hours, minutes and seconds
    m, s = divmod(total_seconds, 60)
    h, m = divmod(m, 60)

    # Format time as h:m:s - 30 seconds of text
    transcript = transcript + "{:0>2d}:{:0>2d}:{:0>2d} {}\n".format(h, m, s, t['text'])

print(transcript)

with open("transcripts/transcript.txt", "w") as f:
    f.write(transcript)
```
